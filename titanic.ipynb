```python
normalizer = pp.Normalizer('l1')
age_array = titanic[['Age']]
age_normalized = normalizer.fit_transform(age_array)
```

```python
titanic.loc[titanic['Sex'] == 'male', 'Sex' ] = 1
titanic.loc[titanic['Sex'] == 'female', 'Sex' ] = 0
```

```python
age_binarizer = binarizer.fit_transform(ages_array)
```

```python
titanic['age_bin'] = age_binarizer
# creates a new column ('age_bin') to house the now binarized ages
```

```python
titanic= titanic.dropna()
```

```python
EmbarkedColumnDummy = pd.get_dummies(titanic['Embarked'])
SexColumnDummy = pd.get_dummies(titanic['Sex'])
```

```python
titanic = pd.concat((titanic, EmbarkedColumnDummy, SexColumnDummy), axis = 1)
```

```python
titanic.head()
```

```python
titanic = titanic.drop(['Sex','Embarked'], axis = 1)
```

```python
X = titanic.values
y = titanic['Survived'].values
```

```python
import numpy as np
X = np.delete(X,1, axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(X_train,y_train)
nb_clf.score(X_test,y_test)
```

```python
titanic.info()
```

```python
from sklearn import ensemble
gb_clf = ensemble.GradientBoostingClassifier()
gb_clf.fit(X_train,y_train)
gb_clf.score(X_test,y_test)
```
